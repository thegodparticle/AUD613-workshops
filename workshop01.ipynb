### <font color="#4338ca"><strong>Part 0: The Setup</strong></font> <span style="color: #64748b; font-size: 0.8em; font-weight: normal;">(Do not skip this)</span>

```python
# =====================================================================
# PART 0: THE SETUP  (run this first)
# =====================================================================

# 1. Import the toolboxes
import numpy as np                                   # math: sines, pi, arrays
import matplotlib.pyplot as plt                      # graphs
from IPython.display import Audio, display           # play sound in the browser
from scipy.signal import butter, lfilter             # digital filters (Part 7)

# 2. Plot defaults for easy viewing
plt.rcParams["figure.figsize"] = (10, 4)
plt.rcParams["font.size"] = 12
plt.rcParams["axes.grid"] = True
plt.rcParams["grid.alpha"] = 0.3

# 3. Global sampling rate (CD quality). 44,100 samples per second.
SR = 44100

# ---------------------------------------------------------------------
# HELPER 1: play a sound safely (normalize so it never clips your ears)
# ---------------------------------------------------------------------
def play_sound(signal, rate=SR, normalize=True):
    """Play audio inline.
    normalize=True  (default): scale to a safe, consistent level so nothing clips.
    normalize=False: play the TRUE amplitude, so loudness differences are audible
                     (use this when the amplitude itself is the point, e.g. Lab 1.2).
    """
    if normalize:
        peak = np.max(np.abs(signal))
        signal = signal / peak * 0.9 if peak > 0 else signal   # avoid divide-by-zero (silence)
        display(Audio(signal, rate=rate))
    else:
        # IPython's Audio normalizes by default; turn that off and keep within [-1, 1]
        display(Audio(np.clip(signal, -1.0, 1.0), rate=rate, normalize=False))

# ---------------------------------------------------------------------
# HELPER 2: a clean time vector of `dur` seconds at sampling rate `sr`
# ---------------------------------------------------------------------
def time_vector(dur=1.0, sr=SR):
    """Return an array of sample times: [0, 1/sr, 2/sr, ...]."""
    return np.arange(int(sr * dur)) / sr

# ---------------------------------------------------------------------
# HELPER 3: the decibel (dB SPL) <-> pressure-amplitude converters
# Reference pressure in air is 20 micropascals (the quietest audible tone).
# ---------------------------------------------------------------------
P_REF = 20e-6   # 20 micropascals

def db_to_amplitude(db_spl):
    """dB SPL -> PEAK amplitude of a sine with that RMS level."""
    rms = P_REF * 10 ** (db_spl / 20.0)
    return rms * np.sqrt(2)          # peak = rms * sqrt(2) for a sine

def amplitude_to_db(signal):
    """Any signal -> its level in dB SPL (from its RMS)."""
    rms = np.sqrt(np.mean(signal ** 2))
    return 20 * np.log10(rms / P_REF) if rms > 0 else -np.inf

# ---------------------------------------------------------------------
# HELPER 4: quantize a signal to n_bits (used in Part 6)
# ---------------------------------------------------------------------
def quantize(signal, n_bits):
    """Round a [-1, 1] signal onto 2**n_bits evenly spaced levels."""
    levels = 2 ** n_bits
    stepped = np.round((signal + 1) * (levels / 2))     # shift to 0..levels, round
    return (stepped / (levels / 2)) - 1                 # shift back to -1..1

# ---------------------------------------------------------------------
# HELPER 5: plot a magnitude spectrum (used in Parts 5-7)
# ---------------------------------------------------------------------
def plot_spectrum(signal, sr=SR, fmax=4000, color="#4338ca", title="Spectrum"):
    """Show which frequencies are present, using the FFT."""
    spectrum = np.abs(np.fft.rfft(signal))
    freqs = np.fft.rfftfreq(len(signal), 1 / sr)
    plt.plot(freqs, spectrum, color=color)
    plt.xlim(0, fmax)
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Magnitude")
    plt.title(title)

print("Setup complete. Libraries imported and helpers ready.")
```

### <font color="#0284c7"><strong>Part 1: Synthesize a Pure Tone</strong></font>

<div style="font-size: 1.05em; color: #374151; margin-bottom: 20px;">
  In lecture we wrote the pure tone as a continuous function of time. A computer cannot store a smooth curve, so it stores a long list of numbers: the height of the wave at evenly spaced instants. This is <strong>discretization</strong>.
</div>

<div style="background-color: #f0f9ff; border: 1px solid #bae6fd; border-left: 6px solid #0284c7; padding: 20px; border-radius: 8px; margin-bottom: 20px;">
  <strong style="color: #0369a1; font-size: 1.1em;">Continuous vs. Discrete</strong>
  <br><br>
  <strong>On paper (continuous):</strong>
  <div style="text-align:center; font-size:1.2em; margin:12px 0; color:#0c4a6e;"><i>x</i>(<i>t</i>) = <i>A</i>&nbsp;sin(2&pi;<i>f t</i> + &phi;)</div>
  <strong>In the computer (discrete):</strong> we sample time at a fixed rate <i>f</i><sub>s</sub>, replacing <i>t</i> with <i>n</i>/<i>f</i><sub>s</sub>:
  <div style="text-align:center; font-size:1.2em; margin:12px 0; color:#0c4a6e;"><i>x</i>[<i>n</i>] = <i>A</i>&nbsp;sin(2&pi;<i>f</i> &middot; <i>n</i>/<i>f</i><sub>s</sub> + &phi;)</div>
  with <i>n</i> = 0, 1, 2, &hellip; The four ingredients are the same as the lecture: amplitude <i>A</i>, frequency <i>f</i>, phase &phi;, and now the sampling rate <i>f</i><sub>s</sub>.
</div>

<div style="padding: 15px; border: 1px solid #bfdbfe; background-color: #eff6ff; border-radius: 8px; border-left: 6px solid #3b82f6;">
  <strong style="color: #1d4ed8;">Action:</strong> Run the cell to generate a 440 Hz reference tone (musical note A4). Read the comments to see how the time vector is built, look at the zoomed plot, then press Play.
</div>

```python
# --- 1. CHOOSE THE FOUR INGREDIENTS ---
freq  = 440.0     # frequency f, in Hz (440 Hz = musical note A4)
amp   = 0.5       # amplitude A (linear, 0 to 1)
phase = 0.0       # phase phi, in radians
dur   = 1.0       # duration, in seconds

# --- 2. BUILD THE TIME VECTOR (discretization) ---
t = time_vector(dur)                 # [0, 1/SR, 2/SR, ...], SR = 44100 per second

# --- 3. SYNTHESIZE: compute the sine for all 44,100 samples at once ---
x = amp * np.sin(2 * np.pi * freq * t + phase)

# --- 4. PLOT a 20 ms window so we can see individual cycles ---
plt.plot(t, x, color="#0284c7", linewidth=2)
plt.xlim(0, 0.02)                    # 0 to 20 milliseconds
plt.ylim(-1.1, 1.1)
plt.title(f"Time domain: {freq:.0f} Hz pure tone")
plt.xlabel("Time (s)"); plt.ylabel("Amplitude")
plt.show()

# --- 5. LISTEN ---
print(f"Generated {len(x)} samples at {SR} Hz. Press play:")
play_sound(x)
```

### <font color="#7c3aed"><strong>Part 2: Measure Amplitude and the Decibel</strong></font>

<div style="font-size: 1.05em; color: #374151; margin-bottom: 20px;">
  "Amplitude" is not one number. Depending on what we measure, we use different definitions, and the one that tracks loudness is RMS. Then we convert RMS to the unit you will use every clinical day: the decibel.
</div>

<div style="display: grid; grid-template-columns: 1fr 1fr; gap: 12px;">
  <div style="background-color: #f5f3ff; padding: 14px; border-radius: 8px; border-left: 5px solid #7c3aed;"><strong style="color: #5b21b6;">Peak (<i>A</i><sub>p</sub>)</strong><br>Maximum displacement from zero.</div>
  <div style="background-color: #eff6ff; padding: 14px; border-radius: 8px; border-left: 5px solid #2563eb;"><strong style="color: #1e40af;">Peak-to-Peak (<i>A</i><sub>pp</sub>)</strong><br>Trough to crest, the full swing.</div>
  <div style="background-color: #fff7ed; padding: 14px; border-radius: 8px; border-left: 5px solid #ea580c;"><strong style="color: #9a3412;">Rectified average</strong><br>Average of the absolute value.</div>
  <div style="background-color: #f0fdf4; padding: 14px; border-radius: 8px; border-left: 5px solid #16a34a;"><strong style="color: #166534;">RMS (<i>A</i><sub>rms</sub>)</strong><br>Root mean square, the effective level. Drives <strong>loudness</strong> and dB.</div>
</div>

<br>
<div style="padding: 15px; border: 1px solid #c4b5fd; background-color: #f5f3ff; border-radius: 8px; border-left: 6px solid #8b5cf6;">
  <strong style="color: #5b21b6;">Action:</strong> Run the cell. It computes six measures for one pure sine. Memorize the factors: <strong>RMS = 0.707 x Peak</strong>, mean square = RMS<sup>2</sup> = 0.5 x Peak<sup>2</sup>, full-wave average = 0.637 x Peak, half-wave average = 0.318 x Peak, peak-to-peak = 2 x Peak.
</div>

```python
# --- Generate a simple 5 Hz wave (low freq just so the plot is readable) ---
t = time_vector(1.0)
y = 1.0 * np.sin(2 * np.pi * 5 * t)

# --- The amplitude measures ---
peak        = np.max(np.abs(y))
peak2peak   = np.max(y) - np.min(y)
full_avg    = np.mean(np.abs(y))             # full-wave rectified average (flip negatives up)
half_avg    = np.mean(np.clip(y, 0, None))   # half-wave average (keep only the positive half)
mean_square = np.mean(y ** 2)                # average of the squared wave
rms         = np.sqrt(mean_square)           # root of the mean square

print("--- AMPLITUDE REPORT (peak = 1.0 sine) ---")
print(f"Peak:              {peak:.4f}")
print(f"Peak-to-Peak:      {peak2peak:.4f}   (= 2 x Peak)")
print(f"Full-wave average: {full_avg:.4f}   (= 2A/pi  = 0.637 x Peak)")
print(f"Half-wave average: {half_avg:.4f}   (= A/pi   = 0.318 x Peak)")
print(f"Mean square:       {mean_square:.4f}   (= A^2/2 = 0.500 x Peak^2)")
print(f"RMS:               {rms:.4f}   (= A/sqrt(2) = 0.707 x Peak)")

# --- Draw the metrics on the waveform ---
plt.plot(t[:8820], y[:8820], color="#cbd5e1", linewidth=3, label="waveform")
plt.axhline(peak,     color="#2563eb", linestyle="--", label=f"Peak ({peak:.3f})")
plt.axhline(rms,      color="#16a34a", linestyle="-",  label=f"RMS ({rms:.3f})")
plt.axhline(full_avg, color="#ea580c", linestyle=":",  label=f"Full-wave avg ({full_avg:.3f})")
plt.axhline(half_avg, color="#a855f7", linestyle="-.", label=f"Half-wave avg ({half_avg:.3f})")
plt.title("Ways to measure amplitude")
plt.xlabel("Time (s)"); plt.legend(loc="upper right")
plt.show()
```

### <font color="#0ea5e9"><strong>Part 3: Superposition, Interference, and Beats</strong></font>

<div style="font-size: 1.05em; color: #374151; margin-bottom: 20px;">
  Real sounds are mixtures. The rule for combining them is the simplest in all of acoustics: at every instant, you <strong>add</strong> the waves together. Waves do not bounce off each other, they pass through and their heights sum.
</div>

<div style="background-color: #f0f9ff; border: 1px solid #bae6fd; border-left: 6px solid #0ea5e9; padding: 20px; border-radius: 8px;">
  <strong style="color: #0284c7; font-size: 1.1em;">The Principle of Superposition</strong>
  <div style="text-align:center; font-size:1.2em; margin:12px 0; color:#0c4a6e;"><i>y</i>(<i>t</i>) = <i>y</i><sub>1</sub>(<i>t</i>) + <i>y</i><sub>2</sub>(<i>t</i>)</div>
  When two waves line up (in phase) they reinforce. When they are half a cycle apart they cancel. When their frequencies are close but not equal, the reinforcement and cancellation alternate, and we hear <strong>beats</strong>.
</div>

<br>
<div style="padding: 15px; border: 1px solid #bfdbfe; background-color: #eff6ff; border-radius: 8px; border-left: 6px solid #3b82f6;">

### <font color="#4338ca"><strong>Part 4: Modulation (AM and FM)</strong></font>

<div style="font-size: 1.05em; color: #374151; margin-bottom: 20px;">
  In Part 3 we <em>added</em> waves. Now one wave will <strong>control</strong> another. A slow modulator shapes a fast carrier. If it shapes the carrier's loudness, that is Amplitude Modulation. If it shapes the carrier's pitch, that is Frequency Modulation.
</div>

<div style="background-color: #eef2ff; border: 1px solid #c7d2fe; border-left: 6px solid #4f46e5; padding: 20px; border-radius: 8px;">
  <strong style="color: #3730a3; font-size: 1.1em;">Amplitude Modulation</strong>
  <div style="text-align:center; font-size:1.2em; margin:12px 0; color:#3730a3;"><i>y</i><sub>AM</sub>(<i>t</i>) = [ 1 + <i>m</i>&nbsp;sin(2&pi;<i>f</i><sub>m</sub><i>t</i>) ] &middot; sin(2&pi;<i>f</i><sub>c</sub><i>t</i>)</div>

</div>

<br>
<div style="padding: 15px; border: 1px solid #bfdbfe; background-color: #eff6ff; border-radius: 8px; border-left: 6px solid #3b82f6;">

### <font color="#86198f"><strong>Part 5: Complex Sounds, Fourier, and Noise</strong></font>

<div style="font-size: 1.05em; color: #374151; margin-bottom: 20px;">
  Here is the deepest idea in the course: <strong>every sound is a stack of sine waves</strong>. Fourier proved that any repeating wave, no matter how jagged, equals a sum of sines at the harmonics: <i>f</i>, 2<i>f</i>, 3<i>f</i>, &hellip; Change the recipe of amplitudes and you change the timbre.
</div>

<div style="background-color: #fdf4ff; border: 1px solid #f5d0fe; border-left: 6px solid #86198f; padding: 20px; border-radius: 8px;">
  <strong style="color: #701a75; font-size: 1.1em;">Additive synthesis: the sawtooth recipe</strong>
  <div style="text-align:center; font-size:1.2em; margin:12px 0; color:#701a75;"><i>y</i>(<i>t</i>) = &Sigma;<sub><i>n</i>=1</sub><sup><i>N</i></sup> (1/<i>n</i>)&nbsp;sin(2&pi;<i>n f t</i>)</div>
  Every harmonic is present, each one quieter by 1/<i>n</i>. We will build it harmonic by harmonic and watch a smooth set of sines turn into a sharp sawtooth. The <strong>spectrum</strong> (from the FFT) shows the recipe directly.
</div>

<br>
<div style="padding: 15px; border: 1px solid #f5d0fe; background-color: #fdf4ff; border-radius: 8px; border-left: 6px solid #a21caf;">
  <strong style="color: #86198f;">Action:</strong> Run to build a sawtooth from 15 harmonics. Left: the waveform. Right: its spectrum, one spike per harmonic.
</div>

```python
# --- DEMO: build a sawtooth by adding harmonics ---
t = time_vector(1.0)
f0 = 220.0           # fundamental frequency
N  = 15              # number of harmonics

saw = np.zeros_like(t)
for n in range(1, N + 1):
    saw += (1.0 / n) * np.sin(2 * np.pi * n * f0 * t)

fig, ax = plt.subplots(1, 2, figsize=(12, 4))
ax[0].plot(t, saw, color="#86198f"); ax[0].set_xlim(0, 0.02)
ax[0].set_title(f"Waveform: {N} harmonics of {f0:.0f} Hz"); ax[0].set_xlabel("Time (s)")

spec = np.abs(np.fft.rfft(saw)); freqs = np.fft.rfftfreq(len(saw), 1/SR)
ax[1].stem(freqs, spec, basefmt=" "); ax[1].set_xlim(0, 3500)
ax[1].set_title("Spectrum: one spike per harmonic"); ax[1].set_xlabel("Frequency (Hz)")
plt.tight_layout(); plt.show()
print("Playing the synthesized sawtooth:")
play_sound(saw)
```

```python
# --- LAB 5.6: vowels and formants (the vowel triangle) ---
# A vowel is a glottal BUZZ (harmonics of the voice pitch f0) shaped by the
# resonances of the vocal tract, called FORMANTS. The first two formants
# (F1, F2) are enough to tell one vowel from another.

VOWELS = {
    "ee  (/i/, heed)": (270, 2290, 3010),
    "ah  (/a/, hot)":  (730, 1090, 2440),
    "oo  (/u/, who)":  (300,  870, 2240),
}

# --- STUDENT CHANGE HERE ---
vowel = "ee  (/i/, heed)"      # try "ee  (/i/, heed)", "ah  (/a/, hot)", "oo  (/u/, who)"
# ---------------------------

def synth_vowel(formants, f0=120, dur=1.0, bw=90):
    """Glottal buzz (harmonics of f0, 1/n) shaped by formant resonances."""
    t = time_vector(dur); sig = np.zeros_like(t); n = 1
    while n * f0 < 5000:
        fn = n * f0
        resonance = sum(1.0 / (1.0 + ((fn - F) / bw) ** 2) for F in formants)  # peaks at each formant
        sig += (resonance / n) * np.sin(2 * np.pi * fn * t)
        n += 1
    return sig

y = synth_vowel(VOWELS[vowel])

fig, ax = plt.subplots(1, 2, figsize=(12, 4))
spec = np.abs(np.fft.rfft(y)); fr = np.fft.rfftfreq(len(y), 1/SR)
ax[0].plot(fr, spec, color="#86198f"); ax[0].set_xlim(0, 3500)
for F in VOWELS[vowel][:2]:
    ax[0].axvline(F, color="red", linestyle="--", alpha=0.6)
ax[0].set_title(f"Spectrum of {vowel.strip()}  (F1, F2 dashed)"); ax[0].set_xlabel("Frequency (Hz)")

for name, (F1, F2, F3) in VOWELS.items():           # the vowel space: F2 (x) vs F1 (y)
    ax[1].scatter(F2, F1, s=90)
    ax[1].annotate(name.split()[0], (F2, F1), textcoords="offset points", xytext=(6, 6))
ax[1].set_xlabel("F2 (Hz)"); ax[1].set_ylabel("F1 (Hz)"); ax[1].set_title("The vowel triangle")
ax[1].invert_xaxis(); ax[1].invert_yaxis()           # phonetic convention: F2 reversed on x, F1 reversed on y
plt.tight_layout(); plt.show()
print(f"Playing {vowel.strip()} at a voice pitch of 120 Hz:"); play_sound(y)
```

### <font color="#991b1b"><strong>Part 6: Digitize, Sampling, Aliasing, and Quantization</strong></font>

<div style="font-size: 1.05em; color: #374151; margin-bottom: 20px;">
  A computer cannot store a continuous wave. It must solve two problems: how often to sample in time (sampling rate), and how finely to round each value in amplitude (bit depth). Get either wrong and the signal is corrupted in a characteristic way.
</div>

<div style="background-color: #fef2f2; border: 1px solid #fecaca; border-left: 6px solid #b91c1c; padding: 20px; border-radius: 8px;">
  <strong style="color: #991b1b; font-size: 1.1em;">The Nyquist-Shannon Sampling Theorem</strong>
  <div style="text-align:center; font-size:1.2em; margin:12px 0; color:#991b1b;"><i>f</i><sub>s</sub> &gt; 2<i>f</i><sub>max</sub></div>
  You need at least two samples per cycle. Sample too slowly and a high tone disguises itself as a lower one: an <strong>alias</strong>. The ghost appears at |<i>f</i><sub>s</sub> &minus; <i>f</i><sub>in</sub>|.
</div>

<br>
<div style="padding: 15px; border: 1px solid #c7d2fe; background-color: #eef2ff; border-radius: 8px; border-left: 6px solid #4f46e5;">
  <strong style="color: #3730a3;">Action:</strong> Run to sample a 900 Hz tone with a deliberately slow 1000 Hz sampler. The red dots connect into a false, slow wave.
</div>

```python
# --- DEMO: aliasing ---
fs_bad, f_real, dur = 1000, 900, 0.02
t_fine = np.linspace(0, dur, 10000)
t_samp = np.arange(0, dur, 1/fs_bad)

plt.plot(t_fine, np.sin(2*np.pi*f_real*t_fine), color="blue", alpha=0.3, label=f"true {f_real} Hz")
plt.plot(t_samp, np.sin(2*np.pi*f_real*t_samp), "o--", color="red", label="samples + alias")
plt.title(f"Sampling {f_real} Hz at only {fs_bad} Hz"); plt.xlabel("Time (s)")
plt.legend(loc="upper right"); plt.show()

f_alias = abs(fs_bad - f_real)
print(f"Nyquist limit: {fs_bad/2:.0f} Hz. Input {f_real} Hz is too fast.")
print(f"Alias appears at |{fs_bad} - {f_real}| = {f_alias} Hz. Playing the alias:")
play_sound(0.5*np.sin(2*np.pi*f_alias*time_vector(1.0)))
```

```python
# --- LAB 6.4: how many dB is ONE BIT worth? (quantization SNR) ---
# Theory for a full-scale sine: SNR (dB) = 6.02 * n + 1.76
# Each extra bit halves the step size, which is about +6 dB of headroom.

t = time_vector(0.5)
signal = np.sin(2 * np.pi * 440 * t)       # full-scale sine (peak = 1.0)

print(f"{'bits':>4}  {'measured SNR (dB)':>18}  {'theory 6.02n+1.76':>18}")
for n in [2, 4, 6, 8, 12, 16]:
    err = signal - quantize(signal, n)                       # the rounding error
    snr = 20 * np.log10(np.sqrt(np.mean(signal**2)) / np.sqrt(np.mean(err**2)))
    print(f"{n:>4}  {snr:>18.1f}  {6.02*n + 1.76:>18.1f}")
```

  <strong style="color: #9a3412;">Action:</strong> Run to make white noise (all frequencies), then pass it through a 1000 Hz low-pass filter. The spectrum shows everything above the cutoff getting cut.

```python
# --- Filter helpers (built on SciPy's butter, imported in Part 0) ---
def lowpass(data, cutoff, fs=SR, order=6):
    b, a = butter(order, cutoff/(fs/2), btype="low")
    return lfilter(b, a, data)

def bandpass(data, low, high, fs=SR, order=4):
    b, a = butter(order, [low/(fs/2), high/(fs/2)], btype="band")
    return lfilter(b, a, data)

def notch(data, center, fs=SR, half_width=50, order=4):
    b, a = butter(order, [(center-half_width)/(fs/2), (center+half_width)/(fs/2)], btype="bandstop")
    return lfilter(b, a, data)

# --- DEMO: low-pass white noise at 1000 Hz ---
noise = np.random.uniform(-0.5, 0.5, SR)
filtered = lowpass(noise, 1000)

plt.magnitude_spectrum(noise,    Fs=SR, scale="dB", color="#cbd5e1", alpha=0.6)
plt.magnitude_spectrum(filtered, Fs=SR, scale="dB", color="#ea580c")
plt.xlim(0, 5000); plt.ylim(-100, -20)
plt.title("Low-pass at 1000 Hz: highs are removed"); plt.show()
print("White noise, then low-passed (muffled):")
play_sound(noise[:SR//2]); play_sound(filtered[:SR//2])
```

```python
# --- LAB 7.5: resonance (a peak right at the cutoff) ---

# --- STUDENT CHANGE HERE ---
boost = 4.0      # size of the resonant peak. try 0.0 (flat) vs 6.0 (whistling)
# ---------------------------

cutoff = 1000
noise = np.random.uniform(-0.5, 0.5, SR)
flat = lowpass(noise, cutoff, order=4)
peak = bandpass(noise, cutoff - 50, cutoff + 50) * boost     # emphasize the cutoff band
resonant = flat + peak

plt.magnitude_spectrum(flat,     Fs=SR, scale="dB", color="gray", label="plain low-pass")
plt.magnitude_spectrum(resonant, Fs=SR, scale="dB", color="red",  label="resonant (high Q)")
plt.xlim(0, 3000); plt.ylim(-60, 0); plt.legend()
plt.title("A resonant bump at the cutoff frequency"); plt.show()
print("Plain:");    play_sound(flat[:SR//2])
print("Resonant:"); play_sound(resonant[:SR//2])
```
