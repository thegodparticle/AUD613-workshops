```python
# --- INSTRUCTOR NOTE: Hide this code block in the student version ---

# 1. Import necessary libraries
import numpy as np                  # The math toolbox (for sines, pi, arrays)
import matplotlib.pyplot as plt     # The graphing toolbox
from IPython.display import Audio   # The toolbox to play sound right in the browser

# 2. Set default plot styles for easier viewing
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (12, 4)     # Make graphs wide and short
plt.rcParams['font.size'] = 12               # Make font readable

# 3. A helper function to play sound safely
def play_sound(waveform, sample_rate=44100):
    """
    Normalizes audio to prevent harsh digital clipping and returns an Audio widget.
    """
    # Normalize the audio to fit within -1.0 to 1.0 (safety volume)
    max_val = np.max(np.abs(waveform))
    if max_val > 0:
        normalized_wave = waveform / max_val * 0.9
    else:
        normalized_wave = waveform # Handle silence case

    return Audio(normalized_wave, rate=sample_rate)

  # --- HELPER: Play Sound Wrapper ---
def play_sound(signal, rate):
    display(Audio(signal, rate=rate))

print("Setup complete! Libraries imported and audio engine ready.")
```

### <font color="#0284c7"><strong>Part 1: Generating the Signal (Digital Synthesis)</strong></font>

<div style="font-size: 1.05em; color: #374151; margin-bottom: 20px;">
  In lecture, we defined a pure tone mathematically as a continuous function of time. However, to analyze or synthesize sound in a computer (DSP), we must discretize this function.
  <br><br>
  This process involves two critical parameters:
  <ol>
    <li><strong>Sampling Rate ($f_s$):</strong> The temporal resolution (how many data points per second).</li>
    <li><strong>Bit Depth:</strong> The amplitude resolution (not covered in this specific exercise, but implied by the float precision).</li>
  </ol>
</div>

<div style="background-color: #f0f9ff; border: 1px solid #bae6fd; border-left: 6px solid #0284c7; padding: 20px; border-radius: 8px; margin-bottom: 20px;">
  <strong style="color: #0369a1; font-size: 1.1em;">The Continuous vs. Discrete Model</strong>
  <br><br>
  <strong>Mathematical Definition (Continuous):</strong>
  $$x(t) = A \sin(2\pi f t + \phi)$$
  <br>
  <strong>Computational Definition (Discrete):</strong>
  We replace continuous time $t$ with a discrete sample index $n$:
  $$x[n] = A \sin\left(2\pi f \frac{n}{f_s} + \phi\right)$$
  <br>
  Where $n$ is an integer array $[0, 1, 2, ... N-1]$.
</div>

<div style="padding: 15px; border: 1px solid #bfdbfe; background-color: #eff6ff; border-radius: 8px; border-left: 6px solid #3b82f6;">
  <strong style="color: #1d4ed8;">Action Required:</strong>
  <br>
  1. <strong>Analyze</strong> the code below to understand how the time vector `t` is constructed using `np.linspace`.
  <br>
  2. <strong>Run</strong> the cell to generate a standard 440 Hz reference tone.
  <br>
  3. <strong>Observe</strong> the zoomed-in plot. Notice that while it looks smooth, it is actually a sequence of connected points.
</div>

```python
# --- 1. DEFINE PARAMETERS ---
# Signal Specifications
sr = 44100          # Sampling Rate (fs): 44.1 kHz (Standard Audio)
duration = 1.0      # Duration in seconds

freq = 440.0        # Frequency (f): 440 Hz (A4)
amp = 0.5           # Amplitude (A): 0.5 (Linear gain, -6 dBFS approx)
phase = 0.0         # Phase (phi): 0 radians

# --- 2. DISCRETIZATION (The Time Vector) ---
# We create a vector 't' representing the exact time of each sample.
# Size of array = sr * duration = 44100 samples.
# np.linspace generates linearly spaced numbers.
num_samples = int(sr * duration)
t = np.linspace(0, duration, num_samples)

# --- 3. SYNTHESIS ---
# We use NumPy's vectorized operations to calculate the sine value for all 44100 points at once.
# x[n] = A * sin(2*pi*f*t + phi)
y = amp * np.sin(2 * np.pi * freq * t + phase)

# --- 4. VISUALIZATION (Temporal Domain) ---
plt.figure(figsize=(10, 4))

# We plot the signal. Note: We restrict the view to a few cycles.
# A 440 Hz wave has a period of ~2.27 ms.
# Showing 20ms allows us to see ~9 cycles clearly.
plt.plot(t, y, color='#0284c7', linewidth=2, label=f'{freq} Hz Tone')

plt.xlim(0, 0.02) # Zoom: 0 to 20 ms
plt.ylim(-1.1, 1.1) # Headroom for visualization
plt.title(f"Time Domain Representation: {freq} Hz")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude (normalized)")
plt.legend(loc='upper right')
plt.grid(True, alpha=0.3)
plt.show()

# --- 5. SONIFICATION ---
print(f"Generated {len(y)} samples at {sr} Hz.")
play_sound(y, sr)
```

### <font color="#7c3aed"><strong>Part 2: The Anatomy of Amplitude</strong></font>

<div style="font-size: 1.05em; color: #374151; margin-bottom: 20px;">
  In physics, "Amplitude" is not just one number. Depending on what we are measuring (voltage, pressure, or loudness), we use different definitions.
</div>

<div style="display: grid; grid-template-columns: 1fr 1fr; gap: 15px;">

  <div style="background-color: #f5f3ff; padding: 15px; border-radius: 8px; border-left: 5px solid #7c3aed;">
    <strong style="color: #5b21b6;">1. Instantaneous Amplitude</strong>
    <br>
    The value at a single specific moment in time.
    <br>
    <em style="color: #6b7280; font-size: 0.9em;">(Input for the graph)</em>
  </div>

  <div style="background-color: #eff6ff; padding: 15px; border-radius: 8px; border-left: 5px solid #2563eb;">
    <strong style="color: #1e40af;">2. Peak Amplitude ($A_p$)</strong>
    <br>
    The maximum positive displacement from zero.
    <br>
    <em style="color: #6b7280; font-size: 0.9em;">(Used in physics equations)</em>
  </div>

  <div style="background-color: #fff7ed; padding: 15px; border-radius: 8px; border-left: 5px solid #ea580c;">
    <strong style="color: #9a3412;">3. Peak-to-Peak ($A_{pp}$)</strong>
    <br>
    The total distance from the bottom trough to the top peak.
    <br>
    <em style="color: #6b7280; font-size: 0.9em;">(Used in engineering/voltage)</em>
  </div>

  <div style="background-color: #f0fdf4; padding: 15px; border-radius: 8px; border-left: 5px solid #16a34a;">
    <strong style="color: #166534;">4. RMS Amplitude ($A_{rms}$)</strong>
    <br>
    "Root Mean Square." The effective average power.
    <br>
    <em style="color: #6b7280; font-size: 0.9em;">(Used for <strong>Loudness</strong> & dB)</em>
  </div>

</div>

<br>

<div style="padding: 15px; border: 1px solid #c4b5fd; background-color: #f5f3ff; border-radius: 8px; border-left: 6px solid #8b5cf6;">
  <strong style="color: #5b21b6;">Action Required:</strong>
  <br>
  Run the code below. It calculates all four values for a standard sine wave.
  <br>
  Note that for a pure sine wave, <strong>RMS is always 0.707 $\times$ Peak</strong>.
</div>

```python
# --- 1. GENERATE A STANDARD WAVE ---
# We use a simple 1.0 amplitude wave to make the math easy to see
amp_peak = 1.0
t_p2 = np.linspace(0, 1.0, 44100)
y_p2 = amp_peak * np.sin(2 * np.pi * 5 * t_p2) # 5 Hz for visibility

# --- 2. CALCULATE THE METRICS ---

# A. Instantaneous (Let's pick a random point, e.g., sample #1000)
inst_amp = y_p2[1000]

# B. Peak Amplitude (Max value)
# We use abs() because the peak could be negative
calc_peak = np.max(np.abs(y_p2))

# C. Peak-to-Peak (Max - Min)
calc_p2p = np.max(y_p2) - np.min(y_p2)

# D. Average (Rectified)
# We must take the absolute value first, otherwise the average is 0!
calc_avg = np.mean(np.abs(y_p2))

# E. RMS (Root Mean Square)
# 1. Square it -> 2. Mean it -> 3. Root it
calc_rms = np.sqrt(np.mean(y_p2**2))

# --- 3. PRINT RESULTS ---
print(f"--- AMPLITUDE REPORT ---")
print(f"1. Instantaneous (at index 1000): {inst_amp:.4f}")
print(f"2. Peak Amplitude:                {calc_peak:.4f}")
print(f"3. Peak-to-Peak:                  {calc_p2p:.4f}  (Should be 2.0)")
print(f"4. Average (Rectified):           {calc_avg:.4f}  (Theoretical: 0.637)")
print(f"5. RMS Amplitude:                 {calc_rms:.4f}  (Theoretical: 0.707)")
print("-" * 30)

# --- 4. VISUALIZE THE DIFFERENCE ---
plt.figure(figsize=(10, 5))
plt.plot(t_p2[:8820], y_p2[:8820], color='#cbd5e1', linewidth=3, label='Waveform') # Grey wave

# Draw lines for the metrics
plt.axhline(calc_peak, color='#2563eb', linestyle='--', linewidth=2, label='Peak (1.0)')
plt.axhline(calc_rms, color='#16a34a', linestyle='-', linewidth=2, label=f'RMS ({calc_rms:.3f})')
plt.axhline(calc_avg, color='#ea580c', linestyle=':', linewidth=2, label=f'Average ({calc_avg:.3f})')

plt.title("Visualizing Amplitude Metrics")
plt.xlabel("Time (s)")
plt.legend(loc='upper right')
plt.grid(True, alpha=0.3)
plt.show()
```

### <font color="#0ea5e9"><strong>Part 3: Synthesis (Adding Waves)</strong></font>

<div style="font-size: 1.05em; color: #374151; margin-bottom: 20px;">
  So far, we have only looked at single, pure sine waves. But real-world sounds (speech, music, noise) are complex.
  <br><br>
  How do we build a complex sound? We use the <strong>Principle of Linear Superposition</strong>.
</div>

<div style="background-color: #f0f9ff; border: 1px solid #bae6fd; border-left: 6px solid #0ea5e9; padding: 20px; border-radius: 8px;">
  <strong style="color: #0284c7; font-size: 1.1em;">The Rule of Addition</strong>
  <br><br>
  To combine two sounds, you simply <strong>ADD</strong> their values at every point in time.
  <br><br>
  $$y_{total}(t) = y_1(t) + y_2(t)$$
  <br>
  Waves do not "bounce" off each other. They pass right through, adding their heights together as they go.
</div>

```python
# --- DEMO: THE PRINCIPLE OF SUPERPOSITION ---
# We will create two waves and add them together.

sr = 44100
t = np.linspace(0, 1.0, sr)

# 1. Create Ingredients
# Wave A: Low and Slow (2 Hz)
freq_a = 5.0
y_a = 1.0 * np.sin(2 * np.pi * freq_a * t)

# Wave B: High and Fast (20 Hz)
freq_b = 50.0
y_b = 0.5 * np.sin(2 * np.pi * freq_b * t)

# 2. The "Mixing Board" (Addition)
y_mix = y_a + y_b

# 3. Visualize
plt.figure(figsize=(10, 8))

# Top Plot: Wave A
plt.subplot(3, 1, 1)
plt.plot(t, y_a, color='blue')
plt.title(f"Wave A: {freq_a} Hz")
plt.grid(True, alpha=0.3)

# Middle Plot: Wave B
plt.subplot(3, 1, 2)
plt.plot(t, y_b, color='orange')
plt.title(f"Wave B: {freq_b} Hz")
plt.grid(True, alpha=0.3)

# Bottom Plot: The Sum
plt.subplot(3, 1, 3)
plt.plot(t, y_mix, color='green', linewidth=2)
plt.title("The Sum (A + B)")
plt.ylabel("Amplitude")
plt.xlabel("Time (s)")
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Listen (We speed up the freq for audio so you can hear it)
print("Playing the Mix (shifted to audible frequencies)...")
# (Shifting to 440 + 880 for audibility)
audio_demo = 0.5 * np.sin(2*np.pi*440*t) + 0.3 * np.sin(2*np.pi*880*t)
Audio(audio_demo, rate=sr)
```
