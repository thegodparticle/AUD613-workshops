```python
# --- INSTRUCTOR NOTE: Hide this code block in the student version ---

# 1. Import necessary libraries
import numpy as np                  # The math toolbox (for sines, pi, arrays)
import matplotlib.pyplot as plt     # The graphing toolbox
from IPython.display import Audio   # The toolbox to play sound right in the browser

# 2. Set default plot styles for easier viewing
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (12, 4)     # Make graphs wide and short
plt.rcParams['font.size'] = 12               # Make font readable

# 3. A helper function to play sound safely
def play_sound(waveform, sample_rate=44100):
    """
    Normalizes audio to prevent harsh digital clipping and returns an Audio widget.
    """
    # Normalize the audio to fit within -1.0 to 1.0 (safety volume)
    max_val = np.max(np.abs(waveform))
    if max_val > 0:
        normalized_wave = waveform / max_val * 0.9
    else:
        normalized_wave = waveform # Handle silence case

    return Audio(normalized_wave, rate=sample_rate)
 # --- HELPER: Play Sound Wrapper ---
def play_sound(signal, rate):
    display(Audio(signal, rate=rate))

print("Setup complete! Libraries imported and audio engine ready.")
```

### <font color="#2563eb"><strong>Part 1: Amplitude Modulation (AM)</strong></font>

<div style="font-size: 1.05em; color: #374151; margin-bottom: 20px;">
  We have learned to add waves. Now we will learn to <strong>multiply</strong> them.
  <br>
  Modulation is the process of using one signal (the Modulator) to change the properties of another (the Carrier).
</div>

<div style="background-color: #eff6ff; border: 1px solid #bfdbfe; border-left: 6px solid #2563eb; padding: 20px; border-radius: 8px; margin-bottom: 20px;">
  <strong style="color: #1e40af; font-size: 1.1em;">The Recipe: Multiplication</strong>
  <br><br>
  Unlike adding waves (Superposition), AM involves multiplying a fast wave by a slow wave:
  $$y_{AM}(t) = \underbrace{[1 + m \sin(2\pi f_m t)]}_{\text{The Envelope}} \times \underbrace{\sin(2\pi f_c t)}_{\text{The Carrier}}$$
  <br>
  Where:
  <ul style="margin-bottom: 0;">
    <li><strong style="color: #1e40af;">$f_c$</strong> = Carrier Frequency (The Pitch)</li>
    <li><strong style="color: #1e40af;">$f_m$</strong> = Modulator Frequency (The Speed of the "Wobble")</li>
    <li><strong style="color: #1e40af;">$m$</strong> = Modulation Depth (0.0 to 1.0)</li>
  </ul>
</div>

<div style="padding: 15px; border: 1px solid #c7d2fe; background-color: #eef2ff; border-radius: 8px; border-left: 6px solid #4f46e5;">
  <strong style="color: #3730a3;">Action Required:</strong>
  <br>
  1. <strong>Run</strong> the code below to generate an AM signal.
  <br>
  2. <strong>Listen</strong> to the result. Do you hear two tones? Or one tone that changes loudness?
  <br>
  3. <strong>Observe</strong> the graph. Notice how the fast blue wave is "trapped" inside the slow envelope.
</div>

```python
# --- 1. DEFINE PARAMETERS ---
sr = 44100
t = np.linspace(0, 1.0, sr)

# Carrier: The Pitch (Fast)
fc = 440.0
carrier = np.sin(2 * np.pi * fc * t)

# Modulator: The Rhythm (Slow)
fm = 5.0
m = 1.0 # Depth (100%)
modulator = m * np.sin(2 * np.pi * fm * t)

# --- 2. CREATE THE SIGNAL (MULTIPLICATION) ---
# We add 1.0 to the modulator so it oscillates between 0 and 2,
# rather than -1 and 1. This prevents "Phase Reversal" (negative amplitude).
envelope = 1.0 + modulator
y_am = envelope * carrier

# --- 3. VISUALIZE ---
plt.figure(figsize=(10, 6))

# Top Plot: The Envelope
plt.subplot(2, 1, 1)
plt.plot(t, envelope, color='#eab308', linewidth=2, label='Envelope (Modulator)')
plt.plot(t, np.zeros_like(t), 'k--', alpha=0.3) # Zero line
plt.xlim(0, 0.5) # Show half second
plt.title(f"The Envelope (Rate: {fm} Hz)")
plt.legend(loc='upper right')
plt.grid(True, alpha=0.3)

# Bottom Plot: The AM Signal
plt.subplot(2, 1, 2)
plt.plot(t, y_am, color='#2563eb', alpha=0.8, label='AM Signal')
# Overlay the envelope for visual proof
plt.plot(t, envelope, color='#eab308', linestyle='--', alpha=0.8, label='Envelope Limit')
plt.xlim(0, 0.5)
plt.ylim(-2.5, 2.5)
plt.title("The Result: Carrier trapped in Envelope")
plt.xlabel("Time (s)")
plt.legend(loc='upper right')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# --- 4. LISTEN ---
print("Playing 440 Hz with 5 Hz Tremolo...")
play_sound(y_am, sr)
```

### <font color="#9333ea"><strong>Part 2: Frequency Modulation (FM)</strong></font>

<div style="font-size: 1.05em; color: #374151; margin-bottom: 20px;">
  In Amplitude Modulation, we changed the <em>size</em> of the wave. In Frequency Modulation, we change the <em>speed</em> of the wave.
</div>

<div style="background-color: #faf5ff; border: 1px solid #e9d5ff; border-left: 6px solid #9333ea; padding: 20px; border-radius: 8px; margin-bottom: 20px;">
  <strong style="color: #6b21a8; font-size: 1.1em;">The Recipe: Nested Functions</strong>
  <br><br>
  FM is non-linear because the modulator is <strong>inside</strong> the sine function:
  $$y_{FM}(t) = A_c \sin(2\pi f_c t + \underbrace{\beta \sin(2\pi f_m t)}_{\text{Phase Modulation}})$$
  <br>
  Where:
  <ul style="margin-bottom: 0;">
    <li><strong style="color: #6b21a8;">$f_c$</strong> = Carrier Frequency (Center Pitch)</li>
    <li><strong style="color: #6b21a8;">$f_m$</strong> = Modulator Frequency (Speed of Vibrato)</li>
    <li><strong style="color: #6b21a8;">$\beta$</strong> = Modulation Index (Depth of Vibrato / Complexity Knob)</li>
  </ul>
</div>

<div style="padding: 15px; border: 1px solid #c4b5fd; background-color: #f5f3ff; border-radius: 8px; border-left: 6px solid #7c3aed;">
  <strong style="color: #5b21b6;">Action Required:</strong>
  <br>
  1. <strong>Run</strong> the code below to generate an FM signal.
  <br>
  2. <strong>Listen</strong> to the "Vibrato" effect.
  <br>
  3. <strong>Observe</strong> the graph. Notice that the amplitude is perfectly flat, but the wave gets "squished" and "stretched" horizontally.
</div>

```python
# --- 1. DEFINE PARAMETERS ---
sr = 44100
t = np.linspace(0, 1.0, sr)

# Carrier: 440 Hz
fc = 440.0
amp = 0.5

# Modulator: 5 Hz Vibrato
fm = 5.0

# Modulation Index (Beta): How "wide" the pitch swing is
beta = 2.0

# --- 2. CREATE THE SIGNAL (NESTED) ---
# Note: The modulator is INSIDE the sine function.
# It adds to the phase angle, speeding up and slowing down the rotation.
y_fm = amp * np.sin(2 * np.pi * fc * t + beta * np.sin(2 * np.pi * fm * t))

# --- 3. VISUALIZE ---
plt.figure(figsize=(10, 6))

# Top Plot: The Modulator (Phase Offset)
plt.subplot(2, 1, 1)
plt.plot(t, beta * np.sin(2 * np.pi * fm * t), color='#a855f7', linewidth=2)
plt.xlim(0, 0.5)
plt.title(f"The Modulator (Rate: {fm} Hz, Beta: {beta})")
plt.ylabel("Phase Shift (rad)")
plt.grid(True, alpha=0.3)

# Bottom Plot: The FM Signal
plt.subplot(2, 1, 2)
plt.plot(t, y_fm, color='#9333ea')
plt.xlim(0, 0.5)
plt.ylim(-1.0, 1.0)
plt.title("The Result: Variable Density (Squished vs Stretched)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude (Constant!)")
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# --- 4. LISTEN ---
print("Playing 440 Hz with Vibrato...")
play_sound(y_fm, sr)
```

### <font color="#b91c1c"><strong>Part 3: Sampling & Aliasing</strong></font>

<div style="font-size: 1.05em; color: #374151; margin-bottom: 20px;">
  A digital computer cannot record a continuous wave. It takes snapshots.
  <br>
  The speed of these snapshots is the <strong>Sampling Rate ($f_s$)</strong>.
</div>

<div style="background-color: #fef2f2; border: 1px solid #fecaca; border-left: 6px solid #b91c1c; padding: 20px; border-radius: 8px; margin-bottom: 20px;">
  <strong style="color: #991b1b; font-size: 1.1em;">The Nyquist-Shannon Sampling Theorem</strong>
  <br><br>
  To correctly record a frequency $f$, you must capture at least 2 samples per cycle.
  <br><br>
  $$f_s > 2 \cdot f_{max}$$
  <br>
  If you violate this rule, the high frequency will "disguise" itself as a low frequency. This ghost signal is called an <strong>Alias</strong>.
</div>

<div style="padding: 15px; border: 1px solid #c7d2fe; background-color: #eef2ff; border-radius: 8px; border-left: 6px solid #4f46e5;">
  <strong style="color: #3730a3;">Action Required:</strong>
  <br>
  1. <strong>Run</strong> the code below. It simulates a "Slow Sampler" (1000 Hz).
  <br>
  2. <strong>Listen</strong> to the result. We try to play a 900 Hz tone, but we hear something much lower.
  <br>
  3. <strong>Observe</strong> the graph. See how the red dots (samples) connect to form a false, slow wave.
</div>

```python
# --- 1. DEFINE PARAMETERS ---
# We simulate a "Bad" Digital System
fs_bad = 1000       # Sampling Rate: 1000 Hz
nyquist = fs_bad / 2 # Limit: 500 Hz

# The Signal we WANT to record
freq_real = 900     # 900 Hz (Above the limit!)
duration = 0.02     # Short view (20ms)

# --- 2. GENERATE ---
# A. The "Truth" (High Resolution view for reference)
t_analog = np.linspace(0, duration, 10000)
y_analog = np.sin(2 * np.pi * freq_real * t_analog)

# B. The "Digital Capture" (Low Resolution)
# We take samples only at the bad sampling rate
t_digital = np.arange(0, duration, 1/fs_bad)
y_digital = np.sin(2 * np.pi * freq_real * t_digital)

# --- 3. VISUALIZE ---
plt.figure(figsize=(10, 5))

# Plot the True Wave (Faint Blue)
plt.plot(t_analog, y_analog, color='blue', alpha=0.3, label='True Analog Wave (900 Hz)')

# Plot the Samples (Red Dots)
plt.plot(t_digital, y_digital, 'o', color='red', label='Digital Samples')

# Plot the "Ghost" Wave (The computer connects the dots)
plt.plot(t_digital, y_digital, '--', color='red', alpha=0.8, linewidth=2, label='The Alias (100 Hz)')

plt.title(f"Aliasing: Recording {freq_real} Hz at {fs_bad} Hz Sample Rate")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend(loc='upper right')
plt.grid(True, alpha=0.3)
plt.show()

# --- 4. LISTEN ---
# Calculate the Alias Frequency: |fs - f_in|
f_alias = abs(fs_bad - freq_real)
print(f"Sampling Rate: {fs_bad} Hz")
print(f"Input Signal:  {freq_real} Hz")
print(f"Nyquist Limit: {nyquist} Hz (Signal is too fast!)")
print(f"Resulting Alias: {f_alias} Hz")

print(f"\nPlaying the ALIAS (What the computer actually recorded)...")
# We generate the alias sound using standard audio rate so you can hear what the "Bad System" saved.
t_play = np.linspace(0, 1.0, 44100)
y_play = 0.5 * np.sin(2 * np.pi * f_alias * t_play)
play_sound(y_play, 44100)
```

### <font color="#b91c1c"><strong>Part 4: Quantization (Bit Depth)</strong></font>

<div style="font-size: 1.05em; color: #374151; margin-bottom: 20px;">
  Sampling solved the Time problem. Now we must solve the Amplitude problem.
  <br>
  Computers cannot store infinitely precise numbers (e.g., 0.33333...). They must <strong>round</strong> the amplitude to the nearest available level.
</div>

<div style="background-color: #fef2f2; border: 1px solid #fecaca; border-left: 6px solid #b91c1c; padding: 20px; border-radius: 8px; margin-bottom: 20px;">
  <strong style="color: #991b1b; font-size: 1.1em;">The "Staircase" Effect</strong>
  <br><br>
  <strong>Bit Depth ($n$):</strong> Determines the number of vertical "steps" available to describe the wave's height.
  <br>
  $$\text{Levels} = 2^n$$
  <br>
  <ul style="margin-bottom: 0;">
    <li><strong>3-bit:</strong> $2^3 = 8$ levels (Blocky, Distorted)</li>
    <li><strong>16-bit (CD):</strong> $2^{16} = 65,536$ levels (Smooth, Clean)</li>
  </ul>
</div>

<div style="padding: 15px; border: 1px solid #c7d2fe; background-color: #eef2ff; border-radius: 8px; border-left: 6px solid #4f46e5;">
  <strong style="color: #3730a3;">Action Required:</strong>
  <br>
  1. <strong>Run</strong> the code below to hear "Low Bit" audio.
  <br>
  2. <strong>Listen</strong> for the "hiss" or "fuzz" in the background. This is <strong>Quantization Noise</strong>.
  <br>
  3. <strong>Observe</strong> the graph. See how the smooth blue wave gets forced into the blocky red steps.
</div>

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio, display


# --- 1. DEFINE PARAMETERS ---
sr = 44100
t = np.linspace(0, 1.0, sr)
freq = 200 # Low frequency to see the steps clearly
amp = 0.9

# Generate High-Res Analog-like Wave (Float 64)
y_analog = amp * np.sin(2 * np.pi * freq * t)

# --- 2. QUANTIZATION FUNCTION ---
def quantize(signal, n_bits):
    """
    Rounds a signal to a specific number of bits.
    1. Scale signal from [-1, 1] to [0, 2^n - 1]
    2. Round to nearest integer (The Quantization Step)
    3. Scale back to [-1, 1]
    """
    levels = 2 ** n_bits
    # Shift to 0..1, scale to levels, round, scale back, shift back
    y_q = np.round((signal + 1) * (levels / 2))
    y_q = (y_q / (levels / 2)) - 1
    return y_q

# Create a Low-Bit Version (3-bit = 8 levels)
n_bits_demo = 3
y_quantized = quantize(y_analog, n_bits_demo)

# --- 3. VISUALIZE ---
plt.figure(figsize=(10, 5))

# Plot "Analog" (Blue)
plt.plot(t[:1000], y_analog[:1000], color='blue', alpha=0.3, linewidth=3, label='Original (Analog)')

# Plot "Digital" (Red Steps)
plt.step(t[:1000], y_quantized[:1000], color='red', where='mid', label=f'{n_bits_demo}-Bit Quantized')

plt.title(f"Quantization: {n_bits_demo}-Bit ({2**n_bits_demo} Levels)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend(loc='upper right')
plt.grid(True, alpha=0.3)
plt.show()

# --- 4. LISTEN ---
# We calculate the "Error" (The Noise)
error_signal = y_analog - y_quantized

print(f"Playing {n_bits_demo}-Bit Audio (Hear the hiss/distortion)...")
play_sound(y_quantized, sr)

print("\nPlaying the QUANTIZATION NOISE only (The difference)...")
# We amplify the noise slightly so you can hear it clearly
play_sound(error_signal * 2.0, sr)
```

### <font color="#ea580c"><strong>Part 5: Digital Filtering</strong></font>

<div style="font-size: 1.05em; color: #374151; margin-bottom: 20px;">
  We have learned how to analyze sound (FFT) and synthesize sound (Oscillators). Now we will learn how to <strong>sculpt</strong> sound.
  <br>
  A filter is an algorithm that selectively removes or boosts specific frequencies.
</div>

<div style="background-color: #fff7ed; border: 1px solid #fed7aa; border-left: 6px solid #ea580c; padding: 20px; border-radius: 8px; margin-bottom: 20px;">
  <strong style="color: #9a3412; font-size: 1.1em;">The "Magic Trick": Frequency Domain Filtering</strong>
  <br><br>
  Filtering in the time domain is hard (Convolution). Filtering in the frequency domain is easy (Multiplication).
  <br><br>
  <strong>The Recipe:</strong>
  <ol>
    <li><strong>Analyze:</strong> Convert signal to spectrum ($X(f)$) via FFT.</li>
    <li><strong>Multiply:</strong> Multiply by the filter shape ($H(f)$). (e.g., multiply high freqs by 0).</li>
    <li><strong>Synthesize:</strong> Convert back to time domain ($y(t)$) via IFFT.</li>
  </ol>
</div>

<div style="padding: 15px; border: 1px solid #c7d2fe; background-color: #eef2ff; border-radius: 8px; border-left: 6px solid #4f46e5;">
  <strong style="color: #3730a3;">Action Required:</strong>
  <br>
  1. <strong>Run</strong> the code below. It generates white noise (all frequencies).
  <br>
  2. <strong>Listen</strong> to the filtered result.
  <br>
  3. <strong>Observe</strong> the graph. See how the "Low Pass" filter cuts off everything above 1000 Hz.
</div>

```python
# --- 1. GENERATE WHITE NOISE ---
sr = 44100
t = np.linspace(0, 1.0, sr)
# White noise has equal energy at ALL frequencies
noise = np.random.uniform(-0.5, 0.5, len(t))

# --- 2. CREATE A FILTER (The "Shape") ---
from scipy.signal import butter, lfilter, freqz

def butter_lowpass(cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    # 'b' and 'a' are the filter coefficients (The "Recipe")
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return b, a

def butter_lowpass_filter(data, cutoff, fs, order=5):
    b, a = butter_lowpass(cutoff, fs, order=order)
    y = lfilter(b, a, data)
    return y

# Apply Low Pass Filter at 1000 Hz
cutoff_demo = 1000
y_filtered = butter_lowpass_filter(noise, cutoff_demo, sr, order=6)

# --- 3. VISUALIZE (Spectrum) ---
plt.figure(figsize=(10, 5))

# Plot Original Noise Spectrum
plt.magnitude_spectrum(noise, Fs=sr, scale='dB', color='#cbd5e1', alpha=0.5, label='Original White Noise')

# Plot Filtered Spectrum
plt.magnitude_spectrum(y_filtered, Fs=sr, scale='dB', color='#ea580c', linewidth=2, label='Filtered Output')

plt.xlim(0, 5000) # Zoom in to see the cut
plt.ylim(-100, -20)
plt.title(f"Low Pass Filter (Cutoff: {cutoff_demo} Hz)")
plt.legend(loc='upper right')
plt.grid(True, alpha=0.3)
plt.show()

# --- 4. LISTEN ---
print("Playing Original White Noise (Hiss)...")
play_sound(noise[:22050], sr) # Play 0.5s

print(f"\nPlaying Filtered Noise (Muffled)...")
play_sound(y_filtered[:22050], sr)
```

```python
# --- LAB 2: THE TELEPHONE EFFECT ---
# GOAL: Isolate the "Mid-Range" frequencies.

# Helper for Bandpass
from scipy.signal import butter, lfilter
def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    y = lfilter(b, a, data)
    return y

# --- ACTION: CHANGE THESE VALUES ---
# Standard Telephone Bandwidth is approx 300 Hz to 3400 Hz.
# Try narrowing it: low=1000, high=2000
low_cut = 300
high_cut = 3400
# -----------------------------------

# We use a voice-like Sawtooth wave this time
t_saw = np.linspace(0, 1.0, sr)
sawtooth = 0.5 * (np.mod(t_saw * 200, 1.0) * 2.0 - 1.0) # 200 Hz Sawtooth

y_bpf = butter_bandpass_filter(sawtooth, low_cut, high_cut, sr, order=4)

print(f"Playing Bandpass ({low_cut}-{high_cut} Hz)...")
play_sound(y_bpf, sr)
```
